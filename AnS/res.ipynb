**Diccionario de variables:**
> Identificación de la jugadora

* `name`: Nombre de la jugadora. Se usa sólo para identificar los resultados.
* `nationality`: Nacionalidad de la jugadora. También usado para el final.
* `age`: Edad de la jugadora.
* `height`: Altura de la jugadora, en centímetros.

> Características generales

* `position`: Posición en la cancha de la jugadora. Agregar significado de las siglas.
* `preferred_foot`: Right or Left. Label encoder is applied as 0 for left and 1 for right.
* `weak_foot`: Represents how well a player uses his weak foot (e.g. left for righties) rated between 1 to 5.
* `work_rate`: Degree of the effort the player puts in terms of attack and defense rated as low, medium and high. This feature is divided into two new features as AttackWorkRate and DefenseWorkRate. Besides, label encoder is applied as 0 for low, 0.5 for medium and 1 for high. `att_work_rate` y `def_work_rate`
* `overall`: Calidad de desempeño general y valor de la jugadora, basado en las habilidades por posición y en su reputación internacional. Valores desde 1 a 99.
* `potential`: Calificación general máxima que se espera que alcance la jugadora en la cima de su carrera. Valores desde 1 a 99.

> Características ofensivas

* `crossing`: Crossing skill of the player rated between 1-99. Cross is a long-range pass from wings to center.
* `finishing`: Finishing skill of the player rated between 1-99. Finishing in football refers to finish an attack by scoring a goal.
* `heading`: Player’s accuracy to pass or shoot by using his head rated between 1-99.
* `short_passing`: Player’s accuracy for short passes rated between 1-99.
* `volleys`: Player’s ability to perform volleys rated between 1-99.

> Características defensivas

* `marking`: Player’s ability to apply strategies to prevent opposing team from taking the ball rated between 1-99. It is a defensive skill.
* `standing_tackle`: Player’s ability to perform tackle (take the ball from the opposite player) while standing rated between 1-99. It is a defensive skill.
* `sliding_tackle`: Player’s ability to perform tackle by sliding rated between 1-99. It is a defensive skill.

> Movimientos

* `acceleration`: Shows how fast a player can reach his maximum sprint speed rated between 1-99.
* `sprint`: Speed rate of the player rated between 1-99.
* `agility`: Gracefulness and quickness of the player while controlling the ball rated between 1-99
* `reactions`: Acting speed of the player to what happens in his environment rated between 1-99.
* `balance`: Player’s ability to remain steady while running, carrying and controlling the ball rated between 1-99.

> Potencia

* `shot_power`: Player’s strength level of shooting the ball rated between 1-99.
* `jumping`: Player’s jumping skill rated between 1-99.
* `stamina`: Player’s ability to sustain his stamina level during the match rated between 1-99. Players with lower stamina get tired fast.
* `strength`: Physical strength of the player rated between 1-99.
* `long_shots`: Player’s accuracy of shoots from long distances rated between 1-99

> Habilidades físicas

* `dribbling`: Dribbling skill of the player rated between 1-99. Dribbling is carrying the ball without losing while moving in one particular direction.
* `curve`: Player’s ability to curve the ball while passing or shooting rated between 1-99.
* `fk_acc`: Player’s accuracy to score free kick goals rated between 1-99.
* `long_passing`: Player’s accuracy for long passes rated between 1-99.
* `ball_control`: Player’s ability to control the ball rated between 1-99.

> Otras destrezas
* `aggression`: Aggression level of the player while pushing, pulling and tackling rated between 1-99
* `interceptions`: Player’s ability to intercept the ball while opposite team’s players are passing rated between 1-99. It is a defensive skill.
* `positioning`: Player’s ability to place himself in the right position to receive the ball or score goals rated between 1-99.
* `vision`:  Player’s mental awareness about the other players in the team for passing rated between 1-99.
* `penalties`: Player’s accuracy to score goals from penalty rated between 1-99
* `composure`: Player’s ability to control his calmness and frustration during the match rated between 1-99
